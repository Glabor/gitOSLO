```python
import os

fileName="sick.bin"#name of file to read

capFile=open(fileName, 'rb')#file to read
f_size=capFile.seek(0, os.SEEK_END)#length of file to read
print(f_size)
capFile.close()
```

```python
# capFile=open(fileName, 'rb')
# for i in range(int(f_size)):
#     print(int.from_bytes(capFile.read(1), byteorder='big', signed=False))
# capFile.close()
with open(fileName, 'rb') as f:
    lines = f.readlines()
    combined_line = b''.join(lines)
    tab_splitted = combined_line.split(b'\t\r\t')[0:]
```

```python
print(tab_splitted)
tl=[]
sl=[]

for t in tab_splitted:
    if len(t)!=1000:
        print(len(t), t)
    else:
        for i in range(0,1000,4):
            tl.append(int.from_bytes(t[i:i+2],byteorder='big', signed=False))
            sl.append(int.from_bytes(t[i+2:i+4],byteorder='big', signed=False))
```

```python
import pandas as pd
df=pd.DataFrame((tl,sl)).transpose()
df.columns=["tl","sl"]
```
