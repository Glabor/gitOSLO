```python
import os

fileName="2000-01-23_08-01-39.bin"#name of file to read

capFile=open(fileName, 'rb')#file to read
f_size=capFile.seek(0, os.SEEK_END)#length of file to read
print(f_size)
capFile.close()
```

```python
# capFile=open(fileName, 'rb')
# for i in range(int(f_size)):
#     print(int.from_bytes(capFile.read(1), byteorder='big', signed=False))
# capFile.close()
with open(fileName, 'rb') as f:
    lines = f.readlines()
    combined_line = b''.join(lines)
    tab_splitted = combined_line.split(b'\t\r\t')[1:]
```

```python
print(tab_splitted)
tl=[]
xl=[]
yl=[]
zl=[]
for t in tab_splitted:
    if len(t)!=13:
        print(len(t), t)
    else:
        tl.append(int.from_bytes(t[0:4],byteorder='big', signed=False))
        xl.append(int.from_bytes(t[4:7],byteorder='big', signed=False))
        yl.append(int.from_bytes(t[7:10],byteorder='big', signed=False))
        zl.append(int.from_bytes(t[10:13],byteorder='big', signed=False))
```

```python
import pandas as pd
df=pd.DataFrame((tl,xl, yl, zl)).transpose()
df.columns=["tl","xl", "yl", "zl"]
```

```python
import plotly.express as px
fig=px.line(df, x="tl", y="xl", title="sick measure", labels="ax", markers=True)
fig.add_scatter(x=df["tl"], y=df["yl"], mode="lines+markers")
fig.add_scatter(x=df["tl"], y=df["zl"], mode="lines+markers")
# fig.update_layout( showlegend=True)
fig.show()
```
